**Step 1: Importing the libraries**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

**Step 2: Importing dataset**

```python
dataset = pd.read_csv('Data.csv')
```

```python
dataset.head()
```

**Step 3: Handling the missing data**

```python
dataset.isnull().sum()
```

```python
dataset
```

```python
#Replacing Nan value in salary with mean
```

```python
dataset["Salary"]=dataset["Salary"].fillna(dataset['Salary'].mean())
dataset
```

```python
dataset["Age"]=dataset["Age"].fillna(dataset['Age'].median())
dataset
```

**Step 4: Encoding categorical data**

```python
from sklearn.preprocessing import OrdinalEncoder, MinMaxScaler
```

```python
# we have two columns with categorical data
```

```python
OE = OrdinalEncoder()
```

```python
dataset['Country'] = OE.fit_transform(dataset[['Country']])
dataset['Purchased'] = OE.fit_transform(dataset[['Purchased']])
```

```python
dataset
```
